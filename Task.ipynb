```python
# Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to
# scrape four attributes:

driver=webdriver.Chrome("chromedriver.exe")
url='https://www.flipkart.com/'
driver.get(url)
time.sleep(1)
search_prod=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_prod.send_keys('sunglasses')
time.sleep(1)
cross_btn=driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
cross_btn.click()
time.sleep(1)
search_btn=driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_btn.click()
time.sleep(1)
```

```python
brand=[]
description=[]
pricetag=[]
discount=[]

for i in range(1,5,1):

    brand_prod=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    brand_prod
    prod_desc=driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
    prod_desc
    price_tag=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    price_tag
    discount_perc=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
    discount_perc
    
    


    for i in range(len(discount_perc)):

        brand.append(brand_prod[i].text)
        try:
            description.append(prod_desc[i].text)
        except:
            description.append("")
        pricetag.append(price_tag[i].text)
       
    
    click_next=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    click_next.click()
    time.sleep(5)
#     break
```

```python
import pandas as pd
data=pd.DataFrame({})
data['brand_prod']=brand[:100]
data['prod_desc']=description[:100]
data['price_tag']=pricetag[:100]
data['discount_perc']=discount[:100]
data
```

```python
# Q8: Scrape data for first 100 sneakers you find when you visit flipkart.com and
# search for “sneakers” in the search field.

driver=webdriver.Chrome("chromedriver.exe")
url='https://www.flipkart.com/'
driver.get(url)
time.sleep(1)

cross_btn=driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
cross_btn.click()
search_prod=driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_prod.send_keys('sneakers')
search_btn=driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_btn.click()
```

```python
Brand=[]
Description=[]
Price=[]
discount=[]
 
for i in range(1,5,1):
    
    brand_tag=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
    brand_tag
    prod_desc=driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
    prod_desc
    price_tag=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
    price_tag
    discount_tag=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
    discount_tag
    
    print(len(brand_tag))
    print(len(prod_desc))
    print(len(discount_tag))
    print(len(price_tag))
    
    
    for i in range(len(brand_tag)):
        try:
            brand.append(brand_tag[i].text)
        except:
            brand.append("")
        try:
            Price.append(price_tag[i].text)
        except:
            Price.append("")    
        try:
            Description.append(prod_desc[i].text)
        except:
            Description.append("")
        try:
            discount.append(discount_tag[i].text)
        except:
            discount.append("")
        
            
    next_btn=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_btn.click()        
    time.sleep(5)
```
